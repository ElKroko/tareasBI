```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Path: taller_integrador.ipynb
# Función para leer el archivo de datos

def xls_to_pd(xls_file):
    df = pd.read_excel(xls_file)
    return df


df = xls_to_pd('Taller.xls').set_index('FECHA VENTA').drop(['Unnamed: 0', "Unnamed: 9"], axis=1)
df.columns
display(df)

## Check duplicates and Nan values

print("Nan values: ", df.isnull().sum().sum())
print("Duplicated values: ", df.duplicated().sum())

## Print all unique values of "MES" column

print("Unique values of 'MES' column: ", df['MES'].unique())


```
