```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Path: taller_integrador.ipynb
# FunciÃ³n para leer el archivo de datos

def xls_to_pd(xls_file):
    df = pd.read_excel(xls_file)
    return df


df = xls_to_pd('Taller.xls').set_index('FECHA VENTA').drop(['Unnamed: 0', "Unnamed: 9"], axis=1)
df.columns
display(df)

## Check duplicates and Nan values

print("Nan values: ", df.isnull().sum().sum())
print("Duplicated values: ", df.duplicated().sum())

## Print all unique values of "MES" column

print("Unique values of 'MES' column: ", df['MES'].unique())







df.to_csv('Taller.csv')
```

```python
## Check value counts of unique values of "DESPACHO" and "NUMERO FACTURA" columns

print("Value counts of 'DESPACHO' column: ", df['DESPACHO'].value_counts())
print("Value counts of 'NUMERO FACTURA' column: ", df['NUMERO FACTURA'].value_counts())

## Print the "DESPACHO" values that have more than 1 value count

display("DESPACHO values that have more than 1 value count: ", df['DESPACHO'].value_counts()[df['DESPACHO'].value_counts() > 1])

## Get the rows that have the value counts of "DESPACHO" column greater than 1

df_despacho = df[df['DESPACHO'].isin(df['DESPACHO'].value_counts()[df['DESPACHO'].value_counts() > 1].index)]



display(df_despacho.set_index("DESPACHO").sort_values('DESPACHO'))
df_despacho.shape

```
